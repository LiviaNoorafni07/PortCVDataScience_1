=> Data Cleaning

```python
#Menghapus Colom Unnamed: 12
df.drop(columns=['Unnamed: 12'], inplace=True)
df
```

```python
# Mengganti Output Yes menjadi 1 dan No menjadi 0
df['Output'] = df['Output'].map({'Yes': 1, 'No': 0})
df
```

```python
# menghitung jumlah baris yang sama dengan tertampil di salah satu kolom
df.groupby(["Monthly Income"]).size()
```

```python
df.columns
```

```python
#mengubah nama kolom
df = df.rename(columns = {'Age': 'Umur',
                            'Gender': 'Jenis Kelamin',
                            'Marital Status': 'Status pernikahan',
                            'Occupation': 'Pekerjaan',
                            'Monthly Income': 'Pendapatan bulanan',
                            'Educational Qualifications': 'Kualifikasi Pendidikan',
                            'Family size': 'Jumlah Anggota Keluarga',
                            'latitude': 'Garis lintang',
                            'longitude': 'Garis Bujur',
                            'Pin code': 'Code Pin',
                            'Feedback': 'Umpan Balik'})
df
```

```python
df.groupby(["Garis lintang"]).size()
```

```python
df.groupby(["Garis Bujur"]).size()
```

```python
# Menggunakan metode unique pada kolom 'Garis lintang' di DataFrame 'df'
unique_values = df['Garis lintang'].unique()

# Mencetak nilai unik
print(unique_values)
```

```python
#mengganti istilah atau string baris di kolom tertentu

df['Jenis Kelamin'] = df['Jenis Kelamin'].replace({'Male': 'Laki-laki', 'Female': 'Perempuan'})
#df['Status pernikahan'] = df['Status pernikahan'].replace({'Married': 'Sudah Menikah', 'Prefer not to say': 'Tidak Diberitahukan', 'Single': 'Lajang'})
#df['Pekerjaan'] = df['Pekerjaan'].replace({'Employee': 'Karyawan', 'House wife': 'Ibu Rumah Tangga', 'Self Employeed': 'Bekerja Sendiri', 'Student': 'Mahasiswa'})
#df['Kualifikasi Pendidikan'] = df['Kualifikasi Pendidikan'].replace({'Graduate': 'Lulus', 'Ph.D': 'phD', 'Post Graduate': 'Pascasarjana', 'School': 'Sekolah', 'Uneducated': 'Tidak berpendidikan'})
#df['Umpan Balik'] = df['Umpan Balik'].replace({'Positive': 'Positif', 'Negative ': 'Negatif'})
df
```

```python
#mengganti float ke float tertentu (nama)

mapping_Pendapatanbulanan = {
    'No Income': '0',
    'Below Rs.10000': '-10000',
    '10001 to 25000': '10001-25000',
    '25001 to 50000': '25001-50000',
    'More than 50000': '50000+'
}

df['Pendapatan bulanan'] = df['Pendapatan bulanan'].map(mapping_Pendapatanbulanan)
df
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Umur'], bins=20, kde=True)
plt.title('Distribusi Usia')
plt.xlabel('Umur')
plt.ylabel('Frekuensi')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Output', y='Pendapatan bulanan', data=df)
plt.title('Distribusi Pendapatan Bulanan berdasarkan Output')
plt.xlabel('Output')
plt.ylabel('Pendapatan bulanan')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Umur', y='Pendapatan bulanan', data=df, hue='Output')
plt.title('Hubungan antara Usia dan Pendapatan Bulanan')
plt.xlabel('Umur')
plt.ylabel('Pendapatan bulanan')
plt.legend(title='Output')
plt.show()
```
